```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi

# Authenticate
api = KaggleApi()
api.authenticate()

# Download the Titanic dataset
dataset_path = os.path.expanduser("./dataset")
os.makedirs(dataset_path, exist_ok=True)
api.competition_download_files('titanic', path=dataset_path)
```

```python
import zipfile

extracted_path = dataset_path + "/extracted"
os.makedirs(extracted_path, exist_ok=True)

# Extract the downloaded zip files
for file_name in os.listdir(dataset_path):
    if file_name.endswith(".zip"):
        with zipfile.ZipFile(os.path.join(dataset_path, file_name), 'r') as zip_ref:
            zip_ref.extractall(extracted_path)

print(f"Data source downloaded and extracted to: {extracted_path}")
```
